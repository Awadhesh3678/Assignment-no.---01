# Question 1: What is NumPy, and why is it widely used in Python?
Answer: NumPy is a Python library used for numerical computations. It supports multi-dimensional arrays and provides fast operations, making it useful for data analysis and scientific tasks.


# Question 2: How does broadcasting work in NumPy?
Answer: Broadcasting lets NumPy perform operations on arrays of different shapes by automatically expanding the smaller one without copying data.


# Question 3: What is a Pandas DataFrame?
Answer: A DataFrame is a 2D table-like data structure in Pandas with labeled rows and columns, ideal for handling structured data.


# Question 4: Explain the use of the groupby() method in Pandas.
Answer: groupby() is used to split data into groups based on a column, then apply functions like sum or mean to each group.


# Question 5: Why is Seaborn preferred for statistical visualizations?
Answer: Seaborn is preferred because it makes beautiful, simple statistical plots with less code and works well with Pandas.


# Question 6: What are the differences between NumPy arrays and Python lists?
Answer: NumPy arrays are more efficient because they store elements of the same type and allow fast mathematical operations. Python lists are more flexible but slower since they can hold mixed data types and don’t support vectorized operations.


# Question 7: What is a heatmap, and when should it be used?
Answer: A heatmap uses colors to show values in a matrix. It's useful for visualizing correlations or data intensity.


# Question 8: What does the term “vectorized operation” mean in NumPy?
Answer: Vectorized operations mean doing calculations on entire arrays at once instead of looping through each element, which makes code faster and simpler.


# Question 9: How does Matplotlib differ from Plotly?
Answer: Matplotlib creates static plots that are simple and customizable, while Plotly produces interactive graphs that users can zoom, hover over, and explore in web browsers.


# Question 10: What is the significance of hierarchical indexing in Pandas?
Answer: Hierarchical indexing allows multiple levels of indexing on rows or columns, making it easier to work with complex datasets by organizing data in a tree-like structure.


# Question 11: What is the role of Seaborn’s pairplot() function?
Answer: pairplot() creates a grid of scatterplots showing relationships between every pair of variables, plus histograms on the diagonal. It’s useful for quickly exploring data.


# Question 12: What is the purpose of the describe() function in Pandas?
Answer: describe() gives a quick summary of statistics like mean, median, min, max, and standard deviation for each numerical column, helping to understand the data at a glance.


# Question 13: Why is handling missing data important in Pandas?
Answer: Missing data can lead to incorrect results or errors during analysis. Pandas lets you detect and handle these missing values to make sure your data is clean and accurate.


# Question 14: What are the benefits of using Plotly for data visualization?
Answer: Plotly lets you create interactive and beautiful charts that can be easily shared online. It supports zooming, hovering, and real-time updates, which makes data exploration more engaging.


# Question 15: How does NumPy handle multidimensional arrays?
Answer: NumPy can create and work with arrays of any number of dimensions, like 2D matrices or 3D tensors. It provides functions to manipulate these arrays efficiently along different axes.


# Question 16: What is the role of Bokeh in data visualization?
Answer: Bokeh is used to make interactive visualizations for the web. It helps create plots that users can interact with, like zooming or selecting data points, right inside browsers.


# Question 17: Explain the difference between apply() and map() in Pandas.
Answer: apply() works on DataFrames or Series to apply a function across rows or columns, while map() is used on a Series for element-wise transformations or mapping values.


# Question 18: What are some advanced features of NumPy?
Answer: NumPy offers advanced indexing, broadcasting, linear algebra operations, FFT, and random number generation, making it powerful for scientific computing.


# Question 19: How does Pandas simplify time series analysis?
Answer: Pandas has built-in date/time support with easy resampling, shifting, and rolling window functions, which makes handling time-indexed data straightforward.


# Question 20: What is the role of a pivot table in Pandas?
Answer: Pivot tables summarize and reshape data by grouping and aggregating, allowing quick insights similar to Excel’s pivot tables.



# Question 21: Why is NumPy’s array slicing faster than Python’s list slicing?
Answer: NumPy arrays store data in continuous memory blocks and use optimized C code for slicing, whereas Python lists are arrays of pointers, making slicing slower.


# Question 22: What are some common use cases for Seaborn?
Answer: Seaborn is used for creating boxplots, violin plots, heatmaps, regression plots, and pairplots, which help visualize statistical relationships.

```python
# Question 1: How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = arr.sum(axis=1)
print(row_sums)
```

```python
# Question 2: Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {'A': [10, 20, 30], 'B': [5, 15, 25]}
df = pd.DataFrame(data)
mean_A = df['A'].mean()
print(mean_A)
```

```python
# Question 3: Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.show()
```

```python
# Question 4: How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris').drop('species', axis=1)
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Question 5: Generate a bar plot using Plotly.
import plotly.express as px
data = {'fruits': ['apple', 'banana', 'orange'], 'count': [10, 15, 7]}
fig = px.bar(data, x='fruits', y='count')
fig.show()
```

```python
# Question 6: Create a DataFrame and add a new column based on an existing column.
import pandas as pd
df = pd.DataFrame({'A': [1, 2, 3]})
df['B'] = df['A'] * 2
print(df)
```

```python
# Question 7: Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print(result)
```

```python
# Question 8: Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
x = [1, 2, 3]
y1 = [2, 4, 6]
y2 = [1, 3, 5]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.show()
```

```python
# Question 9: Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd
df = pd.DataFrame({'A': [5, 10, 15, 20]})
filtered_df = df[df['A'] > 10]
print(filtered_df)
```

```python
# Question 10: Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt
data = sns.load_dataset('iris')['sepal_length']
sns.histplot(data)
plt.show()
```

```python
# Question 11: Perform matrix multiplication using NumPy.
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print(result)
```
